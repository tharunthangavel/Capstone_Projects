## Understanding the domain and significance of each features 

### symboling 
symboling”, corresponds to the insurance risk level of a car. Cars are assigned with a risk factor symbol associated with their price. Then, if an automobile is riskier, this symbol is adjusted by moving it up the scale. A value of +3 indicates that the auto is risky, -3 that it is probably pretty safe. Value ranges from -3 to +3.(Categorical). 

These value will significantly affect reselling value of a car. However, these values are insignificant with respect to the price of a new car.

### normalised losses 
symboling”, corresponds to the insurance risk level of a car. Cars are assigned with a risk factor symbol associated with their price. Then, if an automobile is riskier, this symbol is adjusted by moving it up the scale. A value of +3 indicates that the auto is risky, -3 that it is probably pretty safe. Value ranges from -3 to +3.(Categorical). 

These value will significantly affect reselling value of a car. However, these values are insignificant with respect to the price of a new car.

### make 
Brand name of a car has a significant impact on its price such that cars from certain reputed brands may have a comparatively higher price as compared to another car of similar specifications manufactured by other brand 
### fuel-type 
There were only two categories of fuel in 1985, i.e. diesel & gasoline. In general, Diesel Cars were more costly as compared to their petrol counterparts due to the complexity of Fuel Injection System and Gas Exhaust Systems in Diesel Engines as compared to relatively simple design of Gasoline Engines 
### aspiration 
Process of air intake in an internal combustion engine; namely: a. std (standard), b. turbo (with Turbo Charger) 
### num-of-doors 
Number of doors in the car (either a) two door or b) four door) marginally affects the price. While comparing cars built on similar engine and platform, 4 door configurations are expected to be slightly more costly than similar cars with 2 doors configuration. 

### body-style 

Body types of the car impacts the pricing of a car largely. There are five categories in body style in the dataset:  
hardtop, wagon, sedan, hatchback, convertible 

#### Hatchback: 
These are the most economical category of cars. There include various subcategories a) superminis b) small family cars, c) executive cars and some d) sports cars also. Hence, there are prices of the Hatchback

#### Sedan: 
These cars have a closed body (i.e. a fixed metal roof) with the engine, passengers, and cargo (a trunk/boot lid) in separate compartments. Sedans are more spacious and comfortable as compared to hatchback, hence slightly costlier than hatchback models of similar models.   

#### Wagons:
Wagon (a.k.a. station wagon or estate car is an automotive body-style variant of a sedan with its roof extended rearward[1] over a shared passenger/cargo volume with access at the back via a third or fifth door (the liftgate or tailgate), instead of a trunk/boot lid. The body style transforms a standard three-box design into a two-box design — to include an A, B, and C-pillar, as well as a D-pillar. Station wagons can flexibly reconfigure their interior volume via fold-down rear seats to prioritize either passenger or cargo volume. In general, wagons are slightly higher in cost as compared to sedans if built on similar platform and engines. 

#### Hardtop:
These are closed cars characterised by a rigid roof, typically constructed from metal and these cars do not have B-pillar (centre pillar in the body frame of a car) to resemble the design of a convertible. These are often available in two-door configuration. But, a few rare models are also available in four-door configuration. 
The complexity of design considering the styling & safely aspects make the Hardtops one of the expensive configurations may be only second to convertibles.  

#### Convertible:
These open cars are characterised by convertible arrangement, which allows an open-air driving experience, with the ability to provide a roof when required. The methods of retracting and storing the roof vary among car models and manufacturers. Due to the absence of roof and associated centre pillar frame, these car types need special design architecture considering the safety and styling norms. These are the most expensive category of cars in general.

### drive-wheels 
A drive wheels of motor vehcle refers to the wheels that are coupled to the engine through the driveshaft and gears for the purpose of mechanical power transmission thereby causing the vehicle to move. There are four categories : 4wd(Four wheel drive), fwd(Front Wheel drive), rwd(Rear wheel drive)

### engine-location
It is the location of engine with respect to the bosy of the car. It is mainly of two types say: front & rear.

### engine-type
There are different types of engine categories in Cars, such as: dohc, dohcv, l, ohc, ohcf, ohcv, rotor.  
OHC - Dual Overhead Camshaft (widely used in low to moderate power output suitable for small to medium size cars)
DOHC - Dual Overhead Camshaft  (widely used in medium to large size commuter cars)
DOHCV - Dual Overhead Camshaft  V-Twin (widely used in high end performance cars / sports cars for producing better power output as compared to DOHC)
Rotor – Pistonless rotary engine have a characteristic rotary piston mechanism in place of standard linear piston mechanism (i.e. they have No bore or Piston). These type of engines are rare category

### num-of-cylinders
It is the number of cylinders fitted in engine. The engine capacity power output of a vehicle depends upon the no. of cylinders it possessed.(say: eight, five, four, six, three, twelve, two). Large capacity engines are expected to have more number of cylinders and smaller engines have less number of cylinders (in general). 

### fuel-system
The fuel system of the gasoline cars are either based on carburettor (fixed fuel flow rate in barrel say: 1bbl, 2bbl etc.) or have fuel injection systems
Types of Fuel Enjection:
1.	Multi-point fuel injection (mfi or mpfi – both refer to the same mechanism)
2.	sequential port fuel injection (spfi)
3.	single point direct injection (spdi)
whereas diesel cars have fuel injection system  categorized as: indirect diesel injection system or direct diesel injection system.

carburettor based cars are less expensive as compared to cars with fuel injection systems

### normalized-losses 
attribute “normalized-losses” is the relative average loss payment per insured vehicle year. This value is normalized for all autos within a particular size classification (two-door, small, station wagons, sports/specialty, etc…), and represents the average loss per car per year. 

### wheel-base
The wheelbase of any vehicle is the distance between the centre of the front wheels and the centre of the rear wheels

### length 
It is the length of the car 

### width 
It is the width of the car 

### height 
It is the height of the car

### curb-weight 
It is the weight of the car with no passenger or any other material load

### engine-size 
Engine size is the volume of fuel and air that can be pushed through a car's cylinders and is measured in cubic inch. Car engine sizes are normally specified in litres, which is rounded up to the nearest tenth of a litre

### bore 
- It is the internal diameter of the engine cylinder. 
- Big bore size than stroke length increases the engine power output

### stroke 
- It is the distance travel by piston from TDC to BDC.
- A longer stroke than a bore increases the efficiency of the engine.

### compression-ratio 
is the ratio between the volume of the cylinder and combustion chamber in an internal combustion engine at their maximum and minimum values. A high compression ratio is desirable because it allows an engine to extract more mechanical energy from a given mass of air–fuel mixture due to its higher thermal efficiency. This occurs because internal combustion engines are heat engines, and higher compression ratios permit the same combustion temperature to be reached with less fuel, while giving a longer expansion cycle, creating more mechanical power output and lowering the exhaust temperature. The given range is continuous from 7 to 23. (Numerical)
Compression Ration of a car is related to bore and stroke through non-linear relationship.
The compression ratio can be easily found with the following formula:

CR = (Vd + Vc) / Vc
where:

Vd is the displacement volume - the volume that has been swept by a moving piston. It can be calculated according to the cylinder volume formula: Vd = b^2 x s x π / 4.

Vc is the volume above the piston when the piston is at the top dead-centre (its topmost position), also called the compressed volume.

where:
b is the cylinder bore (diameter).
s is the piston stroke length.

Vc = Vchamber + Vpiston + Vgasket + Vclearance

where:

Vchamber is the volume of the combustion chamber at the top of the engine. It's typically measured by pouring a liquid into the chamber and measuring its volume. You can often find it in the manufacturer's description.
Vpiston - if the piston doesn't have a flat top, but is domed or dished, it will take a bit of space (or leave that space free).
Vgasket - the head gasket of thickness t and bore (diameter) g will occupy a volume of Vgasket = g2 * t * π / 4
Vclearance: given by the following formula: 
Vclearance = b^2 x  c x π / 4,  
where: 
b is the cylinder bore (diameter).
c is the deck clearance

### horsepower 
- It corresponds to the maximum power transmitted from the engine to the drive wheels of the car. 
- Higher the value of Horse-Power more would be the acceleration and top speed of the car 
- Compression Ratio in combination with the Engine Size Determine the Horse-Power of the Car. 

### peak-rpm 
- Maximum rotational speed (revolutions per minute) of the crank-shaft of and internal combustion engine  

### city-mpg 
It is the Mileage rating of the car in (miles per gallon) calculated while driving across city under normal riding conditions.

### highway-mpg
It is the Mileage rating of the car in (miles per gallon) calculated while driving on highway under normal riding conditions.

```python
def grid_search_knn(param_grid, knn_tuning_model = KNeighborsRegressor() ):
    
    '''
    
    Function that searches for right hyperparameters of KNN based on R2 score with 5 fold cross validation and 
    returns the same with their performance on train and test data
    
    '''
    
    grid_knn = GridSearchCV(estimator= knn_tuning_model, param_grid = param_grid,scoring='r2', cv=5, n_jobs=-1,verbose=3)
    grid_knn.fit(X_knn_train_scaled, y_train)
    
    print("Best params:", grid_knn.best_params_)    
    print('\n Best cross val R2 score:' , grid_knn.best_score_)
    
    y_predict_test = grid_knn.predict(X_knn_test_scaled)
    y_predict_train = grid_knn.predict(X_knn_train_scaled)
    
    print('\n R2 score for Train data:', r2_score(y_train,y_predict_train))    
    print('\n R2 score for Test data:', r2_score(y_test,y_predict_test))
```
